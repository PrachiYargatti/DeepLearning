# He (Kaiming) Initialization

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

# Define a simple Dense layer with Kaiming (He) uniform initialization
# for ReLU activation functions.
layer_he_uniform = tf.keras.layers.Dense(
    units=10, # Number of output units
    input_shape=(20,), # Number of input units (fan_in = 20)
    kernel_initializer='he_uniform', # Kaiming (He) uniform initialization
    bias_initializer='zeros',
    activation='relu' # Typically used with ReLU activations
)

# Build the layer to create the weights
layer_he_uniform.build(input_shape=(None, 20))

# Get the kernel (weights) of the layer
kernels_he_uniform = layer_he_uniform.kernel

print("Shape of the kernel (weights) with He Uniform:", kernels_he_uniform.shape)
print("\nFirst 5 rows and columns of the initialized kernel weights (He Uniform):")
print(kernels_he_uniform.numpy()[:5, :5])

print("\nMin value of weights (He Uniform):", tf.reduce_min(kernels_he_uniform).numpy())
print("Max value of weights (He Uniform):", tf.reduce_max(kernels_he_uniform).numpy())
print("Mean value of weights (He Uniform):", tf.reduce_mean(kernels_he_uniform).numpy())
print("Standard deviation of weights (He Uniform):", tf.math.reduce_std(kernels_he_uniform).numpy())

# Calculate the expected limit for he_uniform:
fan_in_he = 20 # Only fan_in is used for He initialization
limit_he = tf.math.sqrt(6.0 / fan_in_he).numpy()
print(f"\nExpected limit for he_uniform: +/--{limit_he:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(kernels_he_uniform.numpy().flatten(), kde=True, bins=50, color='skyblue')
plt.title('Distribution of Initialized Weights (He Uniform)')
plt.xlabel('Weight Value')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

# Define a simple Dense layer with Kaiming (He) normal initialization
# for ReLU activation functions.
layer_he_normal = tf.keras.layers.Dense(
    units=10, # Number of output units
    input_shape=(20,), # Number of input units (fan_in = 20)
    kernel_initializer='he_normal', # Kaiming (He) normal initialization
    bias_initializer='zeros',
    activation='relu' # Typically used with ReLU activations
)

# Build the layer to create the weights
layer_he_normal.build(input_shape=(None, 20))

# Get the kernel (weights) of the layer
kernels_he_normal = layer_he_normal.kernel

print("Shape of the kernel (weights) with He Normal:", kernels_he_normal.shape)
print("\nFirst 5 rows and columns of the initialized kernel weights (He Normal):")
print(kernels_he_normal.numpy()[:5, :5])

print("\nMin value of weights (He Normal):", tf.reduce_min(kernels_he_normal).numpy())
print("Max value of weights (He Normal):", tf.reduce_max(kernels_he_normal).numpy())
print("Mean value of weights (He Normal):", tf.reduce_mean(kernels_he_normal).numpy())
print("Standard deviation of weights (He Normal):", tf.math.reduce_std(kernels_he_normal).numpy())

# Calculate the expected standard deviation for he_normal:
fan_in_he = 20
stddev_he = tf.math.sqrt(2.0 / fan_in_he).numpy()
print(f"\nExpected standard deviation for he_normal: {stddev_he:.4f}")
```
