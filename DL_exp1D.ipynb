Xavier (Glorot) Initialization

```python
import tensorflow as tf

# Define a simple Dense layer with Xavier (Glorot) uniform initialization
# For Xavier initialization, the initializer name is 'glorot_uniform' or 'glorot_normal'
# 'glorot_uniform' draws samples from a uniform distribution within [-limit, limit]
# where limit = sqrt(6 / (fan_in + fan_out))
# 'glorot_normal' draws samples from a truncated normal distribution centered on 0
# with stddev = sqrt(2 / (fan_in + fan_out))

# Let's use glorot_uniform
layer = tf.keras.layers.Dense(
    units=10, # Number of output units
    input_shape=(20,), # Number of input units (fan_in = 20, fan_out = 10)
    kernel_initializer='glorot_uniform', # Xavier (Glorot) uniform initialization
    bias_initializer='zeros' # Usually biases are initialized to zeros
)

# Build the layer to create the weights
# This happens implicitly when the layer is called or by calling .build()
layer.build(input_shape=(None, 20)) # (None, 20) means a batch of inputs with 20 features

# Get the kernel (weights) of the layer
kernels = layer.kernel

print("Shape of the kernel (weights):", kernels.shape)
print("\nFirst 5 rows and columns of the initialized kernel weights:")
print(kernels.numpy()[:5, :5])

print("\nMin value of weights:", tf.reduce_min(kernels).numpy())
print("Max value of weights:", tf.reduce_max(kernels).numpy())
print("Mean value of weights:", tf.reduce_mean(kernels).numpy())
print("Standard deviation of weights:", tf.math.reduce_std(kernels).numpy())

# Calculate the expected limit for glorot_uniform:
fan_in = 20
fan_out = 10
limit = tf.math.sqrt(6.0 / (fan_in + fan_out)).numpy()
print(f"\nExpected limit for glorot_uniform: +/-{limit:.4f}")
```
